```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/Users/khnhkd/Desktop/data/sau_xu_ly.csv')
```

```python
df.head(1)
```

```python
df = df.drop(['date', 'furniture', 'directionHome', 'directionBalcony', 'quan', 'huyen', 'quarters', 'images', 'address'], axis=1)
```

```python
df = df.rename({'district': 'address', 
                'pricePerM': 'price', 
                'numBedroom': 'bedroom', 
                'numBathroom': 'toilet',
                'farCenter': 'distance'
               }, axis=1)
```

```python
df = df.reindex(['address', 'bedroom', 'toilet', 'investor', 'acreage', 'price', 'lat', 'long', 'distance'], axis=1)
```

```python
df.bedroom = df.bedroom.astype(np.float64)
df.toilet = df.toilet.astype(np.float64)
```
